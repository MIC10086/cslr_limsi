```python
# imports
import numpy as np
import itertools as it
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
# load data
annot = np.load('../data/processed/DictaSign/annotations.npz', encoding='latin1', allow_pickle=True)
print(annot.files)
nV = len(annot['dataBrut_fls']) # 94
listeV = np.load('../data/processed/DictaSign/list_videos.npy', encoding='latin1')
nTasks = 9

def getSessionVideo(videoIdx):
    return int(listeV[videoIdx][1])
def getTaskVideo(videoIdx):
    return int(listeV[videoIdx][4])
def getSignerVideo(videoIdx):
    return listeV[videoIdx][6:]
def getUniqueSigners():
    signersTotal = []
    for i in range(nV):
        signersTotal.append(getSignerVideo(i))
    return np.unique(np.array(signersTotal))
def valuesCount(vector):
    values = [i[0] for i in it.groupby(list(vector))]
    count = [len(list(group)) for _, group in it.groupby(list(vector))]
    return values, count
def countUnitsFrames(vector):
    totalUnits = 0
    totalFrames = 0
    values, count = valuesCount(vector)
    nUnits = len(values)
    for i in range(nUnits):
        if values[i] > 0:
            totalUnits += 1
            totalFrames += count[i]
    return totalUnits, totalFrames

idxUniqueSigners = getUniqueSigners()
nSigners = idxUniqueSigners.shape[0]

listCategories = ['fls', 'DS', 'PT', 'FBUOY', 'N', 'FS']
idxCategories = [0,9,1,17,18,19]
nCategories = len(listCategories)

listCategoriesDS = ['DSL', 'DSM', 'DSS', 'DSG', 'DSA', 'DST', 'DSX']
idxCategoriesDS = [12,13,14,11,10,15,16]
nCategoriesDS = len(listCategoriesDS)
```

```python
tabCategoriesSignersUnitsFrames = np.zeros((nSigners, nCategories, 2))
tabCategoriesDSSignersUnitsFrames = np.zeros((nSigners, nCategoriesDS, 2))

for iSigner in tqdm(range(nSigners)):
    for iCategory in range(nCategories):
        units = 0
        frames = 0
        for iVideo in range(nV):
            signerIdxVideo = getSignerVideo(iVideo)
            if signerIdxVideo == idxUniqueSigners[iSigner]:
                unitsTmp, framesTmp = countUnitsFrames(annot['dataBrut_'+listCategories[iCategory]][iVideo].flatten())
                units += unitsTmp
                frames += framesTmp
        tabCategoriesSignersUnitsFrames[iSigner, iCategory, 0] = units
        tabCategoriesSignersUnitsFrames[iSigner, iCategory, 1] = frames
    for iCategoryDS in range(nCategoriesDS):
        units = 0
        frames = 0
        for iVideo in range(nV):
            signerIdxVideo = getSignerVideo(iVideo)
            if signerIdxVideo == idxUniqueSigners[iSigner]:
                unitsTmp, framesTmp = countUnitsFrames(annot['dataBrut_'+listCategoriesDS[iCategoryDS]][iVideo].flatten())
                units += unitsTmp
                frames += framesTmp
        tabCategoriesDSSignersUnitsFrames[iSigner, iCategoryDS, 0] = units
        tabCategoriesDSSignersUnitsFrames[iSigner, iCategoryDS, 1] = frames

tabCategoriesTasksUnitsFrames = np.zeros((nTasks, nCategories, 2))
tabCategoriesDSTasksUnitsFrames = np.zeros((nTasks, nCategoriesDS, 2))

for iTask in tqdm(range(nTasks)):
    for iCategory in range(nCategories):
        units = 0
        frames = 0
        for iVideo in range(nV):
            taskIdxVideo = getTaskVideo(iVideo)-1
            if taskIdxVideo == iTask:
                unitsTmp, framesTmp = countUnitsFrames(annot['dataBrut_'+listCategories[iCategory]][iVideo].flatten())
                units += unitsTmp
                frames += framesTmp
        tabCategoriesTasksUnitsFrames[iTask, iCategory, 0] = units
        tabCategoriesTasksUnitsFrames[iTask, iCategory, 1] = frames
    for iCategoryDS in range(nCategoriesDS):
        units = 0
        frames = 0
        for iVideo in range(nV):
            taskIdxVideo = getTaskVideo(iVideo)-1
            if taskIdxVideo == iTask:
                unitsTmp, framesTmp = countUnitsFrames(annot['dataBrut_'+listCategoriesDS[iCategoryDS]][iVideo].flatten())
                units += unitsTmp
                frames += framesTmp
        tabCategoriesDSTasksUnitsFrames[iTask, iCategoryDS, 0] = units
        tabCategoriesDSTasksUnitsFrames[iTask, iCategoryDS, 1] = frames
```
